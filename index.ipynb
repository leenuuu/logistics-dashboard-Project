```python

import pandas as pd

df = pd.read_csv("APL_Logistics.csv", encoding='latin1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Fix column names by replacing spaces with underscores
df.columns = df.columns.str.replace(" ", "_")
```

```python
#handling missing values
df.isnull().sum()
```

```python
#if few missing values, we can drop them 

df = df.dropna()
```

```python
#create a new column 'Delay_Gap' to calculate the difference between actual and scheduled shipping days
df['Delay_Gap'] = df['Days_for_shipping_(real)'] - df['Days_for_shipment_(scheduled)']
```

```python
#create a delay status column based on the delay gap
def delivery_status(x):
    if x > 0:
        return "Delayed"
    elif x == 0:
        return "On-Time"
    else:
        return "Early"

df['Delivery_Status_New'] = df['Delay_Gap'].apply(delivery_status)
```

## KPI Calculation

```python
# On-time %
on_time = (df['Delivery_Status_New'] == "On-Time").mean() * 100

# Average delay
avg_delay = df['Delay_Gap'].mean()

# Late risk
late_risk = df['Late_delivery_risk'].mean()

print(on_time, avg_delay, late_risk)
```
